```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('fashion-mnist_test.csv')
```

```python
train_df.head()
```

```python
train_df.shape
```

```python
train_df.isna().sum()
```

```python
train_df.dropna(inplace = True)
```

```python
train_df.isna().sum()
```

```python
train_df.shape
```

```python
pixels = train_df.drop(columns = 'label')
```

```python
pixels
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
pixels[:] = scaler.fit_transform(pixels)
```

```python
pixels.head()
```

```python
test_df = pd.read_csv('fashion-mnist_train.csv')
```

```python
test_df.head()
```

```python
test_df.shape
```

```python
pixels_test = test_df.drop(columns = 'label')
```

```python
pixels_test.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
pixels_test[:] = scaler.fit_transform(pixels_test)
```

```python
pixels_test.head()
```

```python
pixels.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(500, input_shape=(784,), activation = 'relu'),
    keras.layers.Dense(300, activation = 'relu'),
    keras.layers.Dense(100, activation = 'relu'),
    keras.layers.Dense(10, activation = 'softmax')
])

model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])

model.fit(pixels, train_df['label'], epochs = 15)
```

```python
loss, accuracy = model.evaluate(pixels_test, test_df['label'])
accuracy * 100
```

```python
pred = model.predict(pixels_test)
```
